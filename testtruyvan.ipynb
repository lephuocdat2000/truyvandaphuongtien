```
import glob
import numpy as np
import re
import pandas as pd
df = pd.read_csv('data_csv.csv', encoding='utf-8',names=["Data", "Contents"])
contents = df['Contents'].to_numpy()
# BUOC 1: Đọc file text trong thư mục 'vanban'
#file_paths = glob.glob("C:/tai lieu hoc tap/truy van da phuong tien/vanban/*.txt")

lst_contents = []
dictionary = set()
for content in contents:
    # lst_contents.append(str)
    words = content.replace('"', '').replace('.', '').replace("'","").split()
    lst_contents.append(words)
    dictionary.update(words)
dictionary = list(dictionary)
# Doc noi dung cua tung file txt
# va xay dung tap "dictionary" chua danh sach cac tu
#BUOC 2: Xây dựng inverted files
inv_files = dict()
def build_inverted_files(inv_files):
   for k,word in enumerate(dictionary):
      inv_files[word] = set()
      print(word)
      for i,content in enumerate(lst_contents):
         if word in content:
            count=content.count(word)
            tf=count/len(content)
            tup_word_tf=(i,tf)
            inv_files[word].add(tup_word_tf)
build_inverted_files(inv_files)
#Buoc 3: Tinh toan inv_file va tfidf 
def calcu_word_tfidf(inv_files):
    tf_idf_arr=np.zeros(len(inv_files))
    total_content=len(lst_contents)
    for i,word in enumerate(inv_files):
        k=len(inv_files[word])
        idf=total_content/k
        for tup in inv_files[word]:
            temp=list(tup)
            temp[1]*=idf
            tup=tuple(temp)
calcu_word_tfidf(inv_files)
```
